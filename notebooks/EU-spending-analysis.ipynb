```python
import pandas as pd
```

```python
data = pd.read_csv('../data/tech.csv')
```

```python
data['startDate'] = data['startDate'].map(pd.to_datetime)
data['endDate'] = data['endDate'].map(pd.to_datetime)
```

```python
data['Period'] = [row.strftime('%Y %m') for row in data['startDate']]
```

```python
countries = pd.read_csv('../data/countries.csv')
```

```python
countries.head()
```

```python
table = pd.pivot_table(
    data,
    values='ecMaxContribution',
    index='coordinatorCountry',
    columns='Period',
    aggfunc='sum',
    fill_value=0,
    margins=False,
    dropna=True,
    margins_name='All',
).reset_index()
```

```python
merged = pd.merge(table, countries[['Country','EU','Image URL']], how='left', left_on='coordinatorCountry', right_on='Country').drop('coordinatorCountry', axis=1)
```

```python
#features = merged.loc[:,"ISOeurostats":"Image URL"]
#cumsum = merged.loc[:,"2014 05":"2020 09"].cumsum(axis = 1) 
```

```python
final_table = pd.merge(merged.loc[:,"Country":"Image URL"], 
                       merged.loc[:,"2014 05":"2020 09"].cumsum(axis = 1), 
                       how='inner',left_index=True,right_index=True)
```

```python
final_table.to_csv('../data/table.csv')
```
